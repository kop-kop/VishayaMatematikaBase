```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,4*np.pi,100)
k1=1
k2=13
y1 = [np.cos(i*k1) for i in x]
y2 = [np.cos(i*k2) for i in x]
```

```python
plt.title("y(k,x)=cos(k*x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(x,y1,x,y2)
```
